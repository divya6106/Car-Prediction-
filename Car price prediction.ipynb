```python
car["company"].unique()
```

```python
#feature is x and price is y
X = car.drop(columns='Price')
y = car['Price']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.preprocessing  import OneHotEncoder
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
```

```python
ohe = OneHotEncoder()#liner regression only understand numerical value and easy to understand that why we needd this formation
#Learns patterns (e.g.std, mean, categories).
ohe.fit(X[['name','company','fuel_type']])
#transform() â†’ Applies learned transformations.(convert 0 to 1 )
```

```python
ohe.categories_
```

```python
column_trans = make_column_transformer((OneHotEncoder(categories=ohe.categories_),['name','company','fuel_type']),remainder='passthrough')

""" make_column_transformer() â†’ Applies transformations to specific columns.
    OneHotEncoder() â†’ Converts categorical values into binary columns.
    ['Brand', 'Model', 'Fuel'] â†’ Columns to encode.
    remainder='passthrough' â†’ Keeps all other columns unchanged.ex = price
"""
```

```python
lr = LinearRegression() #Linear Regression is used to predict a continuous value (e.g., house price, salary, car price).
```

```python
pipe = make_pipeline(column_trans,lr)
#Linear Regression is used to predict a continuous value (e.g., house price, salary, car price).
#It first transforms the data (using column_trans).
#Then it trains the Linear Regression model (lr).
#pipeline do transfromation and preict price at a same time that do pipeline 
```

```python
car = car.loc[:, ~car.columns.str.contains('^Unnamed')]  # Remove any unwanted index column

# Now, proceed with pipeline fitting
X = car[['name', 'company', 'year', 'kms_driven', 'fuel_type']]
y = car['Price']  # Assuming 'price' is the target variable

pipe.fit(X, y)
```

```python
pipe.fit(X_train,y_train)
#means that piped data fit that x trained and y trained data
#this code give error because we need to convert data into numeric ordder
```

```python
y_pred = pipe.predict(X_test) #this predict result of based on x_test
```

```python
y_pred
```

```python
r2_score(y_test,y_pred) #y_test = actual value , y_pred = predicted value
#The RÂ² score tells us how well a regression model predicts the actual values.
```

```python
scores = []
for i in range(1000):
    X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2, random_state = i)
    lr = LinearRegression()
    pipe = make_pipeline(column_trans,lr)
    pipe.fit(X_train,y_train)
    y_pred = pipe.predict(X_test)
    scores.append(r2_score(y_test,y_pred))

#r2 score value not fix  train test split they take diffrent data diffrent times
#now we run 1000 time and get highest and maximum r2 score generate 
# The same random numbers are used each time.

    

    
```

```python
import numpy as np
np.argmax(scores) #random state 
```

```python
#check max r2 scores 
scores[np.argmax(scores)]
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2, random_state = np.argmax(scores))
lr = LinearRegression()
pipe = make_pipeline(column_trans,lr)
pipe.fit(X_train,y_train)
y_pred = pipe.predict(X_test)
r2_score(y_test,y_pred)
```
